```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import scipy as sp

df = pd.read_csv(r'ratingdf_likescore.csv')
print(len(df))

df2 = df[['session_id','track_id','rating']]



df2['rating_round'] = df2['rating'].round()
```

```python
BYTES_TO_MB_DIV = 0.000001
def print_memory_usage_of_data_frame(df2):
    mem = round(df.memory_usage().sum() * BYTES_TO_MB_DIV, 3) 
    print("Memory usage is " + str(mem) + " MB")
    
print_memory_usage_of_data_frame(df2)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
import time
from sklearn.cross_validation import train_test_split
for key, item in vector_dict.items():
    
    #print(key)
    
    start = time.time()
    X_train, X_test, y_train, y_test = train_test_split(item[0], item[1], test_size=0.3, random_state=42)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
clf = DecisionTreeClassifier(max_depth=None, max_leaf_nodes=20)

from sklearn import tree
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
y_pred
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print("Precision_score:",metrics.precision_score(y_test, y_pred,average = None))
print("Recall_score:",metrics.recall_score(y_test, y_pred,average = None))
```

```python
from sklearn.ensemble import RandomForestClassifier
clf2 = RandomForestClassifier(max_depth=None, max_leaf_nodes=20)
clf2.fit(X_train, y_train)
y_pred = clf2.predict(X_test)
print(y_pred)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print("Precision_score:",metrics.precision_score(y_test, y_pred,average = None))
print("Recall_score:",metrics.recall_score(y_test, y_pred,average = None))
```
