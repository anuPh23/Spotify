```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
dat = pd.read_csv(r'F:\AppliedDataScience\data\dat.csv')
dat.head()
```

```python
dat.columns
```

```python
plt.figure(figsize=(10, 7))
dat.boxplot(column=['acousticness','beat_strength','bounciness',
                    'danceability', 'energy', 'flatness'],
                        patch_artist='True',vert=True,showmeans=True)
plt.show()
```

```python
plt.figure(figsize=(15, 12))
corr = dat.drop('skip_1',axis = 1).corr()
sns.heatmap(corr[(corr >= 0.5) | (corr <= -0.4)],
                 cmap = 'viridis',annot = True,annot_kws={"size":8},square = True)
plt.show()
```
